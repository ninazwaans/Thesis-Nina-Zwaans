```python
import pandas as pd
from statsmodels.stats.contingency_tables import mcnemar
from itertools import combinations

# Confusion matrices for each fold
conf_matrices = {
    "Fold 1": [[58, 62], [52, 56]],
    "Fold 2": [[42, 78], [62, 46]],
    "Fold 3": [[69, 39], [85, 35]],
    "Fold 4": [[46, 38], [47, 85]],
    "Fold 5": [[66, 66], [58, 38]]
}

# Initialize results list to hold result dictionaries
results_list = []

# Generate all combinations of fold pairs
fold_names = list(conf_matrices.keys())
fold_combinations = combinations(fold_names, 2)  # Creates all possible pairs of folds

for fold1, fold2 in fold_combinations:
    conf_matrix1 = conf_matrices[fold1]
    conf_matrix2 = conf_matrices[fold2]

    # Elements for McNemar's 2x2 table
    b = conf_matrix1[0][1] + conf_matrix2[1][0]
    c = conf_matrix1[1][0] + conf_matrix2[0][1]

    mcnemar_table = [[conf_matrix1[0][0] + conf_matrix2[1][1], b],
                     [c, conf_matrix1[1][1] + conf_matrix2[0][0]]]

    # Perform McNemar test
    result = mcnemar(mcnemar_table, exact=False, correction=True)
    results_list.append({'Comparison': f'{fold1} vs {fold2}',
                         'Statistic': result.statistic,
                         'p-value': result.pvalue})  # Corrected attribute for p-value

# Convert results list to DataFrame
results_df = pd.DataFrame(results_list)

# Display results
print(results_df)
```


In the context of assessing the consistency of classification models across different subsets of data, the McNemar test provides a robust statistical method for comparing the predictive discrepancies between paired models or data splits. This test is particularly useful in scenarios where models are trained and validated on different segments of data, allowing researchers to identify whether differences in performance are due to variations in the data or inherent characteristics of the models themselves.

The McNemar test focuses on the count of instances where one model is correct and the other is incorrect, forming a contingency table with these counts. This approach helps in evaluating the statistical significance of the observed differences, using a chi-squared statistic to determine if the models perform differently on the same set of instances.

The results from applying the McNemar test to various pairs of folds from a dataset are presented in the table. For instance, the comparison between Fold 1 and Fold 3 shows a significant discrepancy in performance (p-value = 0.000364), indicating a statistically significant difference in the error patterns between these two folds. Similarly, the comparison between Fold 2 and Fold 3, and between Fold 3 and Fold 5, also reveals significant differences (p-values of 0.000174 and 0.000764, respectively), suggesting that the model's performance varies notably across these folds. On the other hand, comparisons such as Fold 1 vs Fold 5 and Fold 2 vs Fold 5 do not show significant differences, indicating more consistent performance between these fold pairs.
